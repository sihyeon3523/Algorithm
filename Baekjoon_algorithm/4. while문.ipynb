```python
# 1110
a = int(input())
new_a, new_step = 0, 0

for i in range(len(str(a))):
    new_step += int(str(a)[i])
new_a = int(str(a)[-1]+str(new_step)[-1])
n=1

while True:
    if a == new_a:
        print(n)
        break
    new_step = 0
    for i in range(len(str(new_a))):
        new_step += int(str(new_a)[i])
    new_a = int(str(new_a)[-1]+str(new_step)[-1])
    n +=1
```

```python
# 1110 (//:몫, %: 나머지)
a = int(input())
check = a
new_a, new_step = 0, 0 
n = 0
temp = 0

while True:
    temp = a//10 + a%10
    new_a = (a%10)*10 + temp%10
    n +=1
    a = new_a
    if new_a == check:
        break
print(n)
```
