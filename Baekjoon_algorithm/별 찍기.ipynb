```python
# 2440
def star(n):
    if n == 0:
        return
    print('*'*n)
    star(n-1)
    
N = int(input())
star(N)
```

```python
# 2441
n = int(input())

for i in range(n, 0, -1):
    print(' '*(n-i)+'*'*i) 
    # ','로 연결할 경우, 한칸 뛰고 같이 출력
    # +로 연결할 경우 두개를 붙여서 같이 출력 
```
