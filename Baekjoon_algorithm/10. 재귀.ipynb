```python
# 10872
def factorial(n):
    if n == 1:
        return 1
    return n*factorial(n-1)

print(factorial(5))
```

```python
# 10870
def fibonacci(num):
    if num <=1:
        return num
    return fibonacci(num-1)+fibonacci(num-2)

N = int(input())
print(fibonacci(N))
```

```python
# 2447 
num = int(input())

def star(i, j):
    while (i != 0):
        # 몫이 1인 경우
        if(i%3 == 1 and j %3 == 1): # 나머지가 1? 
            print(' ')
            return None
        # 3으로 나누어서 위의 if문에 걸리면 그 부분도 빈칸 처리 
        i = i//3
        j = j//3
    print('*')
    
for i in range(num):
    for j in range(num):
        star(i, j)
    print('\n')
```
