```python
# 10809
alpha = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q',
         'r','s','t','u','v','w','x','y','z']
loca = [-1]*26
a = list(input())

for i in range(len(a)):
    for j in range(0,26):
        if a[i] == alpha[j]:
            if loca[j] == -1:
                loca[j] = i
        else:
            continue

for i in loca:
    print(i,end=' ')  
```

```python
a = input().upper()
unique_word = list(set(a))

cnt_list = []
for i in unique_word:
    cnt_list.append(a.count(i))
    
if cnt_list.count(max(cnt_list)) >1:
    print('?')
else:
    max_index = cnt_list.index(max(cnt_list))
    print(unique_word[max_index])
```
