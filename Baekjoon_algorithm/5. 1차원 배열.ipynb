```python
# 8958
a = int(input())

for i in range(a):
    b = str(input())
    temp = [0]*len(b)
    
    if b[0] == 'O':
        temp[0] = 1
    else:
        temp[0] = 0
        
    for i in range(1, len(b)):
        if b[i] == 'O':
            temp[i] = temp[i-1]+1
        else:
            temp[i] = 0 
            
    print(sum(temp))
```

```python
n = int(input())
for i in range(n):
    Num = input()
    score = 0
    cnt = 0
    for j in range(len(Num)):
        if Num[j] == 'O':
            cnt += 1
            score += cnt 
        else:
            score += 0
            cnt = 0 
    print(score)
```

```python
# 4344
n = int(input())

for i in range(n):
    n_score = list(map(int, input().split()))
    avg = sum(n_score[1:])/n_score[0]
    cnt = 0
    
    for i in n_score[1:]:
        if i > avg:
            cnt +=1
            
    print("%.3f%%" % round((cnt/n_score[0])*100,3))
```
