```python
# 2750 - 버블정렬
n = int(input())
a = []

for i in range(n):
    a.append(int(input()))

# bubble sort 
for i in range(len(a)):
    for j in range(len(a)):
        if a[i] < a[j]:
            a[i], a[j] = a[j], a[i]

for n in a:
    print(n)
```

```python
# 2750 - 삽입정렬
n = int(input())
a = []

for i in range(n):
    a.append(int(input()))

# insert sort 
for i in range(1, len(a)):
    while (i>0) & (a[i] < a[i-1]):
        a[i], a[i-1] = a[i-1], a[i]
        
        i -= 1

for n in a:
    print(n)
```

```python
# 2751 - 병합정렬
def merge_sort(unsorted_list):
    if len(unsorted_list) <= 1:
        return unsorted_list
    
    mid = len(unsorted_list)//2 # 몫
    left = unsorted_list[:mid]
    right = unsorted_list[mid:]
    
    left1 = merge_sort(left)
    right1 = merge_sort(right)
    return merge(left1, right1)
```

```python
def merge(left, right):
    i = 0
    j = 0
    sorted_list = []
    
    while (i < len(left)) & (j <len(right)):
        if left[i] < right[j]:
            sorted_list.append(left[i])
            i+=1
        else:
            sorted_list.append(right[j])
            j+=1
    
    while (i < len(left)):
        sorted_list.append(left[i])
        i+=1
        
    while (j < len(right)):
        sorted_list.append(right[j])
        j+=1
    
    return sorted_list
```

```python
n = int(input())
unsorted_list = []
for i in range(n):
    unsorted_list.append(int(input()))
    
for i in merge_sort(unsorted_list):
    print(i)
```

```python
def merge_sort(array):
    if len(array)<=1:
        return array
    
    # 재귀함수를 이용해서 끝까지 분할
    mid = len(array)//2
    left = merge_sort(array[:mid])
    right = merge_sort(array[mid:])

    i,j,k = 0,0,0

    # 분할된 배열끼리 비교
    while i<len(left) and j <len(right):
        if left[i]<right[j]:
            array[k] = left[i]
            i += 1
        else:
            array[k] = right[j]
            j += 1
        k+=1
    
    # 먼저 끝났을 때 
    if i==len(left):
        while j < len(right):
            array[k] = right[j]
            j+=1
            k+=1
    elif j==len(right):
        while i < len(left):
            array[k] = left[i]
            i+=1
            k+=1
    return array

# 데이터 입력
n=int(input())
num = []

for _ in range(n):
    num.append(int(input()))

num = merge_sort(num)

for i in num:
    print(i)
```

```python
# 2751 - 힙정렬
# 힙의 성질을 만족하는지 확인, 아래 노드가 위 노드보다 크면 그것을 바꿔줌! 
# 재귀 호출하여 반복!
def heapify(unsorted, index, heap_size):

    largest = index
    left_index = 2*index +1
    right_index = 2*index +2
    # heap_size 와의 비교는 range를 벗어나는지 아닌지 확인하기 위함 
    if (left_index < heap_size) and (unsorted[left_index] > unsorted[largest]):
        largest = left_index
    if (right_index < heap_size) and (unsorted[right_index] > unsorted[largest]):
        largest = right_index
    if largest != index: # 힙의 성질이 만족되도록 바뀌었다면! 값을 바꾼다. 
        unsorted[largest], unsorted[index] = unsorted[index], unsorted[largest]
        heapify(unsorted, largest, heap_size) 
        # unsorted: 값이 바뀐 리스트, largest: 값이 바뀐 인덱스(큰 값), heap_size: n  
```

```python
def heap_sort(unsorted):
    n = len(unsorted)
    for i in range(n//2-1,-1,-1):
        heapify(unsorted, i, n)

    for i in range(n-1, 0, -1):
        unsorted[0], unsorted[i] = unsorted[i], unsorted[0]
        heapify(unsorted, 0, i)
    return unsorted 
```

```python
# 2108
from collections import Counter 
import sys
n = int(sys.stdin.readline())
stat = [] #디폴트 =  한칸만 만든다. 


for i in range(n):
    stat.append(int(sys.stdin.realine()))
    
stat.sort()

#산술평균     
print(round(sum(stat)/n)) 

# 중앙값
print(stat[n//2])

#최빈값 
stat_s = Counter(stat).most_common() # Counter: 요소의 개수를 세준다, most_common:빈도수가 높은 순으로 정렬된 리스트 안의 튜플 리턴 
if len(stat_s) >1:
    if stat_s[0][1] == stat_s[1][1]:
        print(stat_s[1][0])
    else:
        print(stat_s[0][0])
else:
    print(stat_s[0][0])

# 범위
print(stat[-1]-stat[0])
```

```python
# 11650
# 2차원 - x,y 존재
# x 오름차순, x가 같으면 y좌표가 증가하는 순서로 정렬 
# 1열: x좌표 2열: y좌표 
# 2차원 리스트의 요소에 접근하기: 리스트[세로인덱스(행)][가로인덱스(열)]

n = int(input()) # 반복 횟수 인풋 

a = [[0]*2]*n # 2열 n행 2차원 리스트 만들기 

for i in range(n): # n번 for 문 돌기 
    a[i] = list(map(int, input().split())) # input()에서 받은 리스트를 split() 한 것을 int 함수를 적용 시켜 list로 만들기 

a.sort(key=lambda x: (x[0], x[1])) # key 옵션을 사용하여 정렬 진행 x[0]을 기준으로 오름차순
# 첫번째 값이 같다면, 두번째 값을 비교하여 오름차순으로 정렬 (내림차순으로 정렬하고 싶을 경우, -x[0]와 같이 표현)

for i in range(n): 
    print(a[i][0], a[i][1])
```
