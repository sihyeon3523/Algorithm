```python
def solution(array, commands):
    answer = []
    for n in range(len(commands)):
        i = commands[n][0]
        j = commands[n][1]
        k = commands[n][2]
        cut=array[i-1:j]
        cut.sort()
        answer.append(cut[k-1])
    return answer
```
