```python
def solution(record):
    answer = []
    mapping={}
    for x in record:
        if 'Enter' in x or 'Change' in x:
            # id key x.split(' ')[1]  value 
            mapping[x.split(' ')[1]]=x.split(' ')[2]
    for x in record:
        if 'Enter' in x:
            answer.append(mapping.get(x.split(' ')[1]) +" .")
        elif 'Leave' in x:
            answer.append(mapping.get(x.split(' ')[1]) +" .")
    return answer
```
