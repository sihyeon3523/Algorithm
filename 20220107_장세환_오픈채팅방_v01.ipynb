```python
def solution(record):
    answer = []
    mapping={}
    for x in record:
        if 'Enter' in x or 'Change' in x:
            # 유저id를 key로 x.split(' ')[1] 닉네임을 value로 
            mapping[x.split(' ')[1]]=x.split(' ')[2]
    for x in record:
        if 'Enter' in x:
            answer.append(mapping.get(x.split(' ')[1]) +"님이 들어왔습니다.")
        elif 'Leave' in x:
            answer.append(mapping.get(x.split(' ')[1]) +"님이 나갔습니다.")
    return answer
```
